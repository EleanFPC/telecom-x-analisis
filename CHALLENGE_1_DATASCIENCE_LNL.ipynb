**PARTE 1: IMPORTACIÓN DE DATOS**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## 1. ANALISIS DE FACTURACIÓN

```python
#1.1 REVISIÓN DE ESTRUCTURAS

tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    print("="*50)
    print(f"📊 {nombre}")
    print("="*50)

    print("\n🔹 Información general:")
    display(df.info())

    print("\n🔹 Estadísticas descriptivas:")
    display(df.describe())

    print("\n🔹 Valores nulos:")
    display(df.isnull().sum())

    print("\n\n")
```

**1.1 METRICAS COMPARATIVAS ENTRE TIENDAS**

 ================ INGRESOS ===========================

```python
resumen = []

for nombre, df in tiendas.items():
    resumen.append({
        "Tienda": nombre,
        "Ingreso Total": df["Precio"].sum(),
        "Calificación Promedio": df["Calificación"].mean(),
        "Costo Envío Promedio": df["Costo de envío"].mean(),
        "Cantidad Ventas": df.shape[0]
    })

resumen_df = pd.DataFrame(resumen)

resumen_df
```

**====================== CATEGORIAS MÁS VENDIDAS ==================================**

```python
for nombre, df in tiendas.items():
    print("="*60)
    print(f"📊 {nombre} - Categoría más vendida")
    print("="*60)

    categoria_top = df["Categoría del Producto"].value_counts().head(1)
    display(categoria_top)
    print("\n")
```

**============================= VENTAS ENTRE TODAS LAS CATEGORIAS =======================**

```python
for nombre, df in tiendas.items():
    print("="*60)
    print(f"📊 {nombre} - Ventas por Categoría")
    print("="*60)

    display(df["Categoría del Producto"].value_counts())
    print("\n")
```

# **2. RESEÑAS DE CLIENTES: **

```python
# Ya tenemos el promedio, pero ahora vamos a analizar distribución:

for nombre, df in tiendas.items():
    print("="*60)
    print(f"⭐ {nombre} - Distribución de Calificaciones")
    print("="*60)

    display(df["Calificación"].value_counts().sort_index())
    print("\n")
```

## **3. PRODUCTOS MÁS VENDIDOS**

```python
for nombre, df in tiendas.items():
    print("="*60)
    print(f"🏆 {nombre} - Top 5 productos más vendidos")
    print("="*60)

    display(df["Producto"].value_counts().head(5))
    print("\n")
```
